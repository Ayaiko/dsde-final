```python
import pandas as pd
from pipeline.extract import load_traffy_data, load_all_weather_data, load_single_location_weather, load_bangkok_air_quality


df = pd.read_csv('data/processed/traffy_weather_merged.csv')
```

```python
df.head(50)
```

```python
# Replace ' ' (single space) with NaN, then drop those rows
df.replace(' ', pd.NA, inplace=True)
df.dropna(how='any', inplace=True)
print(f"Rows after dropping spaces: {len(df)}")
df.head(10)
```

```python
# Check for ' ' (single space) in any column
space_mask = df.apply(lambda col: col == ' ')
rows_with_spaces = space_mask.any(axis=1)

print(f"Total rows with ' ' in any column: {rows_with_spaces.sum()}")
print(f"\nColumns containing ' ' and their counts:")
print(space_mask.sum())

# Show which rows have spaces
if rows_with_spaces.sum() > 0:
    print(f"\nFirst few rows with spaces:")
    print(df[rows_with_spaces].head())
```
