```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (12,8)
```

```python
# Raw DATA
path = '../data/raw/open-meteo-13.74N100.50E9m.csv'
df_1 = pd.read_csv(path)
print(df_1.shape)
df_1.head()
```

```python
# Nut merge DATA
path = '../data/processed/traffy_merged.csv'
df_2 = pd.read_csv(path)
print(df_2.shape)
df_2.head()
```

```python
df_1.info()
```

```python
df_2.info()
```

```python
df_1['date'] = pd.to_datetime(df_1['time']).dt.date
```

```python
df_1['date'] = pd.to_datetime(df_1['date'])
df_2['date'] = pd.to_datetime(df_2['date'])
```

```python
df_1_daily = df_1.groupby('date').mean(numeric_only=True).reset_index()
```

```python
df_1_daily.head()
```

```python
df_1_daily.shape
```

```python
merged = df_2.merge(df_1_daily, on='date', how='left')
```
